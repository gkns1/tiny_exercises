```python
#slow, huge
def ProdigyFibonacci(n):

	if n <= 1:
		return n

	return ProdigyFibonacci(n - 1) + ProdigyFibonacci(n - 2)


if __name__ == '__main__':

	n = 100
	print("n'th Fibonacci number is", ProdigyFibonacci(n))
```

```python
def fibonacci(n): 
    x = 0
    y = 1
    if n == 0: 
        return 0 
    elif n == 1: 
        return 1 
    else: 
        for i in range(2,n+1): 
            z = x + y 
            x, y = y, z 
        return y
if __name__ == "__main__":
    n = 7777
    print(fibonacci(n))
# O(n) time
# O(1) space
# not the best way to calculate it, but works well enough for this exercise
# or we can "cheat" using the formula Fn = (((5 + 1)/2) ^ n) / 5, but it's fussy in Python
```
